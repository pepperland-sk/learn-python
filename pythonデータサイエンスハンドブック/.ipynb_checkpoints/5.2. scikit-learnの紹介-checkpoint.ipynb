## 5.2.1 scikit-learnのデータ表現

### 5.2.1.1 表としてのデータ

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

行：``サンプル``　→　データセット内の個々のオブジェクト
列：``特徴``　→　各サンプルを定量的に表す観測値


### 5.2.1.2 特徴行列

行列：``特徴行列``

### 5.2.1.3 目的配列

変数yで、1次元、nサンプルの長さを持つ：``目的配列``

```python
%matplotlib inline
import seaborn as sns; sns.set()
sns.pairplot(iris, hue='species', size=1.5);
```

```python
#特徴行列
X_iris = iris.drop('species', axis=1)
X_iris.shape
```

```python
#目的配列
y_iris = iris['species']
y_iris.shape
```

## 5.2.2 scikit-learnの推定器API

### 5.2.2.1 APIの基礎

手順
1. 適切なEstimatorクラスをインポートして、使用するモデルを選択
2. 目的の値でクラスをインスタンス化し、モデルのハイパーパラメータを選択
3. 先に説明したように、データを特徴行列と目的配列として配置
4. インスタンスのfit()メソッドを呼び出し、当てはめを行う。
5. モデルに新しいデータを適用する

### 5.2.2.2 教師あり学習例：単回帰

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
```

1. モデルのクラスを選択する

```python
from sklearn.linear_model import LinearRegression
```

2. モデルのハイパーパラメータを選択する

```python
model = LinearRegression(fit_intercept=True)
model
```

3. データを特徴マトリクスと目的配列に配置する

```python
X = x[:, np.newaxis]
X.shape
```

4. データをモデルに当てはめる

```python
model.fit(X, y)
```

```python
#fit()処理中に学習されたすべてのモデルパラメータには下線がつく
model.coef_
```

```python
model.intercept_
```

5. 未知データのラベルを予測する

```python
xfit = np.linspace(-1, 11)
xfit
```

```python
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
plt.scatter(x, y)
plt.plot(xfit, yfit);
```

### 5.2.2.3 教師あり学習の事例：アイリスの分類

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB #1.モデルクラスの選択
model = GaussianNB() #2. モデルのインスタンス化
model.fit(Xtrain, ytrain) # 3. データの当てはめ
y_model = model.predict(Xtest) #4. 新しいデータに対する予測
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

### 5.2.2.4 教師なし学習の事例：アイリス次元数

```python
from sklearn.decomposition import PCA #1.モデルクラスの選択
model = PCA(n_components=2) #2. モデルのインスタンス化
model.fit(X_iris) # 3. データの当てはめ
X_2D = model.transform(X_iris) #4. 新しいデータに対する予測
```

```python
#結果のプロット
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
sns.lmplot("PCA1", "PCA2", hue='species', data=iris, fit_reg=False)
```

### 5.2.2.5 教師なし学習の例：アイリスクラスタリング

```python
from sklearn.mixture import GaussianMixture #1.モデルクラスの選択
model = GaussianMixture(n_components=3, covariance_type='full') #2. ハイパーパラメータを与えてインスタンス化
model.fit(X_iris) #3. データへの当てはめ（yを指定していない点に注意）
y_gmm = model.predict(X_iris) #4. クラスタのラベルを確認する
```

```python
iris['cluster'] = y_gmm
sns.lmplot("PCA1", "PCA2", data=iris, hue='species', col='cluster', fit_reg=False);
```

## 5.2.3 応用：手書き数字の調査

### 5.2.3.1 数字データのロードと可視化

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(10, 10, figsize=(8, 8),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(digits.target[i]),
            transform=ax.transAxes, color='green')
```

```python
#イメージ内の各ピクセル（8*8 = 64)を特徴として扱う

X=digits.data #特徴行列（特徴マトリクス）
X.shape

```

```python
y = digits.target #目的配列
y.shape
```

```python
#Isomap(多様体学習アルゴリズム)によってデータを2次元に変換
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
iso.fit(digits.data)
data_projected = iso.transform(digits.data)
data_projected.shape
```

```python
plt.scatter(data_projected[:, 0], data_projected[:, 1], c=digits.target, edgecolor='none', alpha=0.5, cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar(label='digit label', ticks=range(10))
plt.clim(-0.5, 9.5)
```

### 5.2.2.3 数字の分類

```python
#データを学習用とテスト用に分割
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB #モデルクラスを選択
model = GaussianNB() #モデルをインスタンス化
model.fit(Xtrain, ytrain) #学習用データにモデルを当てはめる
y_model = model.predict(Xtest) #新しいデータに予測
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
#分類を混同行列で確認
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(ytest, y_model)

sns.heatmap(mat, square=True, annot=True, cbar=False, cmap='RdPu');
plt.xlabel('predicted value')
plt.ylabel('ture value');
```

```python
fig, axes = plt.subplots(10, 10, figsize=(8, 8),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))

test_images = Xtest.reshape(-1, 8, 8)

for i, ax in enumerate(axes.flat):
    ax.imshow(test_images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(y_model[i]),
            transform=ax.transAxes,
            color='green' if (ytest[i] == y_model[i]) else 'red')
```
