## 4.10.1 plt.axes: サブプロットのマニュアル作成

サブプロット：　1つのプロットの中に小さな軸のグループを配置する

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
%matplotlib inline
import numpy as np
```

```python
ax1 = plt.axes()  # standard axes
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2]) #[x軸の位置, y軸の位置, xの範囲, yの範囲]
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                   xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                   ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```

## 4.10.2 plt.subplot: サブプロットの単純なグリッド

```python
#plt.subplotの例
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2,3,i)), fontsize=18, ha='center')
```

```python
#fig.subplots_adjust(hspace=0.4, wspace=0.4): サブプロットの間隔を調整
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
           fontsize=18, ha='center')
```

## 4.10.3 plt.subplots: グリッド全体を一度に作成

```python
#plt.subplots(): x軸とy軸を共有
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row') #sharex, sharey指定によって内側ラベルが自動削除
```

```python
# axes are in a two-dimensional array, indexed by [row, col]
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i, j)),
                      fontsize=18, ha='center')
fig
```

## 4.10.4 plt.GridSpec: より複雑な配置

```python
#2行3列のGridSpecの作成
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
```

```python
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

```python
#正規分布データの作成
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

#GridSpecによる軸配列作成
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

#主軸に散布図をプロット
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

#付随する軸にヒストグラムをプロット
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```
