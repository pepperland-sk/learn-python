`主成分分析(PCA: Principal Component Analysis` 教師なし学習推定器の1つ。基本的に`次元削減`アルゴリズムだが、可視化、ノイズフィルタリング、特徴集出、特徴量エンジニアリングにも使用可能

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

## 5.9.1 主成分分析の基礎

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

x値からy値を予測するのではなく、教師なし学習によりxとyの関係を明らかにする

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)

```

```python
#成分(components)
print(pca.components_)

#因子寄与(explained variance)
print(pca.explained_variance_)
```

「成分」をベクトルの方向、「因子寄与」をベクトルの長さの二乗とおいてベクトルを描画

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

 - ベクトル：データの主軸方向を表す
 - ベクトルの長さ：データの分布を表す歳にその字句がどの程度「重要であるか」（→軸上に投影したデータの分散）を表す
 - 各データポイントの主軸上への投影が「主成分」となる

`アフィン変換` データ軸から主軸への変換。平行移動、回転、均一なスケーリングで構成される

### 5.9.1.1 PCAによる次元削減

```python
#データの次元を縮小し、1次元に変換
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print("original shape:", X.shape)
print("transformed shape:", X_pca.shape)
```

```python
#次元削減の可視化
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.axis('equal');
```

### 5.9.1.2 PCAによる可視化：手書き数字

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
pca = PCA(2) #64次元を2次元に変換
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

```python
plt.scatter(projected[:, 0], projected[:, 1], c=digits.target, edgecolor='none', alpha=0.5, cmap=plt.cm.get_cmap('spectral', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

### 5.9.1.3 成分の意味

ピクセル単位の観点で把握する場合
$$
{\rm image}(x) = x_1 \cdot{\rm (pixel~1)} + x_2 \cdot{\rm (pixel~2)} + x_3 \cdot{\rm (pixel~3)} \cdots x_{64} \cdot{\rm (pixel~64)}
$$

各ピクセルから事前に定義された寄与を含むような基底関数を使用する場合
$$
image(x) = {\rm mean} + x_1 \cdot{\rm (basis~1)} + x_2 \cdot{\rm (basis~2)} + x_3 \cdot{\rm (basis~3)} \cdots
$$

### 5.9.1.4 成分数の選択

`累積因子寄与率` データを記述するのに必要な成分数を特定するのに使用する

```python
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

## 5.9.2 PCAによるノイズフィルタリング

```python
#ノイズなしデータの作成
def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize=(10, 4), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw = dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8), cmap='binary', interpolation='nearest', clim=(0,16))
plot_digits(digits.data)
```

```python
#ノイズの多いデータ
np.random.seed(42)
noisy = np.random.normal(digits.data, 4)
plot_digits(noisy)
```

```python
#PCAを学習させる
pca = PCA(0.50).fit(noisy)
pca.n_components_
```

```python
components = pca.transform(noisy)
filtered = pca.inverse_transform(components)
plot_digits(filtered)
```

## 5.9.3 事例：固有顔画像

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person = 60)
print(faces.target_names)
print(faces.images.shape)
```

```python
#最初の150の成分を確認
from sklearn.decomposition import PCA
pca = PCA(150, svd_solver='randomized')
pca.fit(faces.data)
```

```python
#データ・セットから学習した固有顔画像の可視化
fig, axes = plt.subplots(3, 8, figsize=(9, 4), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(62, 47), cmap='bone')
```

```python
#累積因子寄与率から、150の主成分が分散の90%を占めることがわかる
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
# Compute the components and projected faces
pca = PCA(150, svd_solver='randomized').fit(faces.data)
components = pca.transform(faces.data)
projected = pca.inverse_transform(components)
```

```python
# Plot the results
#3000次元を150次元に削減する
fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(faces.data[i].reshape(62, 47), cmap='binary_r')
    ax[1, i].imshow(projected[i].reshape(62, 47), cmap='binary_r')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('150-dim\nreconstruction');
```
